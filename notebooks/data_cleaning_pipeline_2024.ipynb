```python
### Importação dos packages necesssários

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_path = r'C:\Users\catarina.miranda\PycharmProjects\FicheiroCompras_ML\data\raw\JE_Valérius_Sede_Compras_2024.xlsx'

data = pd.read_excel(data_path)
print('Ficheiro lido')
```

### Análise inicial dos dados

```python
# Primeiro vamos analisar o cabeçalho do ficheiro
data.head(10)
```

```python
# Vamos começar por eliminar colunas e linhas que não nos interessem (A escolha das colunas a remover foi realizada de forma iterativa, ao mesmo tempo em que era realizada a                                                                           análise exploratória que se segue)

remover_colunas = [
    'Unnamed: 0',
    'Alinhamento',
    'Data fisca',
    'Rubrica',
    'Atividade/código',
    'CAPEX/OPEX/Volume Negócios',
    'OBSERVAÇÕES',
    'saldo',
    'Euros  deb',
    'Euros  cre',
    'Centro ana',
    'Doc  n  do',
    'Numero do ',
    'Descritivo']

remover_linhas = [0,1,2,3,4]

dataset = data.drop(columns=remover_colunas)
dataset = dataset.drop(remover_linhas)
```

```python
dataset.head(10)
```

O dataset já não apresenta colunas/linhas que **não** sejam interessantes para a análise.

```python
# Verificação de NAs (Valores Nulos / Vazios)

dataset.isna().sum()
```

A coluna "Afornecedo" possui 3268 NAs e a coluna "N contrib" 110

Abordagem definida para esses casos:

Verificámos que existe vários problemas entre a interação de **N  contrib** e **Afornecedo**. Isso porque, em muitos casos identificados, o contribuinte corresponde ao comprador, e noutros casos corresponde ao vendedor, o que faz com que a sua relação não seja lógica nem padronizada.


- **"Afornecedo"** -> As linhas com NAs serão **eliminadas** (Apesar de serem 3mil linhas, não causam impacto no target como é possível verificar em baixo)

- **"N contrib"** -> As linhas com NAs serão **eliminadas**


```python
#dataset["Afornecedo"] = dataset["Afornecedo"].fillna("Valérius")

print(dataset["Elegível?"].value_counts())

dataset = dataset.dropna(subset=["N  contrib"])
dataset = dataset.dropna(subset=["Afornecedo"])

dataset.isna().sum()
print(dataset["Elegível?"].value_counts())
```

Já não existem valores nulos/inválidos no dataset.

```python
# Vamos transformar a variável binária categórica ('Elegível?') numa variável binária numérica. Assim é mais fácil para o modelo!

dataset['Elegível?'] = dataset['Elegível?'].str.lower().str.strip()
dataset['Elegível?'] = dataset['Elegível?'].map({'sim':1, 'não':0, 'nao':0})

dataset['Elegível?'].value_counts()
```

Também é importante eliminar ruído existente em algumas colunas:

```python
dataset['Doc  inici'].value_counts() # Visto que 14k das linhas possuem PT nesta coluna, tem sentido categorizar as restantes                                         como "Outros"

dataset['Doc  inici'] = dataset['Doc  inici'].apply(lambda x: x if x == 'PT' else 'Outros')

dataset['Doc  inici'].value_counts()
```

```python
# Verificação do dataset para ver se está tudo bem
dataset.head()
```

### Influência de algumas das colunas no target
Vamos analisar algumas colunas que parecem ser das mais interessantes:

```python
colunas_texto = ['Designação Fatura', 'Descritivo', 'Nome da co']

print("=== ANÁLISE DAS COLUNAS DE TEXTO ===")
for coluna in colunas_texto:
    if coluna in dataset.columns:
        print(f"\n--- {coluna} ---")
        print(f"Valores únicos: {dataset[coluna].nunique()}")
        print(f"Missing values: {dataset[coluna].isnull().sum()}")
        print("Exemplos:")
        print(dataset[coluna].dropna().head(3).tolist())
```

```python
nome_co_stats = dataset.groupby('Nome da co').agg({
    'Elegível?': ['sum', 'mean']
}).round(3)

nome_co_stats.sort_values(by=('Elegível?', 'mean'), ascending=False, inplace=True)

nome_co_stats
```

Neste caso é possível observar em que tipos de **"Nome da co"** mais influenciam o **target**, nomeadamente:
- Cons Edf. e O. Const. MN Tx Reduzida;
- Equip. Bás. Máquinas MC Tx Normal
- Resíduos MN Tx Normal
- Cons Veic. Lig Mercadorias MN Tx Normal
- Transp Mercad Normal MN Tx Isento
- Cons Veic. Lig Passageiros MN Base Não Ded
- Cons Veic. Lig Passageiros MN Não Ded Tx Normal
- Transp Mercad Normal MN Tx Normal
- Cons Veic. Pes Mercadorias MN Tx Normal

É de destacar que algumas das colunas possuem uma **elevada taxa de eligibilidade,mesmo aparecendo diversas vezes**.


Vamos fazer o mesmo, mas para a coluna **'Doc  nome1'**:

```python
doc_nome_values = dataset['Doc  nome ']

doc_nome_stats = dataset.groupby('Doc  nome ').agg({
    'Elegível?': ['sum', 'mean']
}).round(3)

doc_nome_stats.sort_values(by=('Elegível?', 'mean'), ascending=False, inplace=True)

print(doc_nome_values)
doc_nome_stats
```

Nenhuma das colunas se destaca relativamente à sua ligação com o target.
